# Hopkin Test

### Cek Kesehatan Cluster (Hopkins Test)
- nilai > 70%, maka layak dipakai
- nilai sekitar 50% maka dianggap random (perlu dimensi reduksi)
- nilai < 30% berarti tidak cocok pakai kmeans, pakai metode yang bisa mengatasi masalah sparse dataset

contoh sparse ada dibawah


```python
#Calculating the Hopkins statistic
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan

def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n)
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)

    rand_X = sample(range(0, n, 1), m)

    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])

    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0

    return H
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
custom_colors = [
    "#FF5733",  # Merah Cerah
    "#33FF57",  # Hijau Cerah
    "#3357FF",  # Biru Cerah
    "#FFFF33",  # Kuning Cerah
    "#FF33A1",  # Magenta Cerah
    "#33FFF7",  # Cyan Cerah
    "#FF8C00",  # Oranye Cerah
    "#8A33FF",  # Ungu Cerah
    "#FF33F6",  # Pink Cerah
    "#C4E17F",  # Hijau Pastel
    "#FFC300",  # Kuning Emas
    "#DAF7A6",  # Hijau Muda
    "#FFC0CB",  # Pink Pastel
    "#900C3F",  # Merah Tua
    "#581845",  # Merah Anggur
    "#DFFF00",  # Kuning Neon
]
# Misalkan data_model sudah terdefinisi sebelumnya
df = data_model[["CAM", "CB", "CDM", "CF", "CM", "GK", "LB", "LM", "LW", "LWB", "RB", "RM", "RW", "RWB", "ST", "Clusters"]]

# Merapihkan data untuk barplot dengan menggunakan melt
df_melt = df.melt(id_vars='Clusters', var_name='Features', value_name='Nilai')

# Filter hanya untuk nilai 1
df_melt_1 = df_melt[df_melt['Nilai'] == 1]

# Fungsi untuk membuat barplot hanya untuk cluster tertentu (misalnya cluster 1)
def plot_categorical_distribution_for_cluster(feature, cluster_number):
    # Filter data hanya untuk cluster yang dipilih
    cluster = df_melt_1[df_melt_1['Clusters'] == cluster_number]

    # Menghitung distribusi kategori unik untuk cluster yang dipilih
    counts = cluster.groupby([feature]).size().reset_index(name='Jumlah')

    # Membuat barplot
    plt.figure(figsize=(10, 6))

    # Menggunakan palet warna cerah
    palette = sns.color_palette(custom_colors, n_colors=len(counts))

    # Membuat barplot untuk cluster tertentu
    bars = plt.bar(counts[feature], counts['Jumlah'], color=palette)

    # Menambahkan keterangan (legend)
    plt.title(f'Feature Positions Distribution in Clusters {cluster_number}')
    plt.xlabel(feature)
    plt.ylabel('amount')

    # Menambahkan keterangan warna
    legend_labels = counts[feature].unique()
    for color, label in zip(palette, legend_labels):
        plt.bar(0, 0, color=color, label=label)  # Tambahkan dummy bar untuk legend

    # Menempatkan legend di luar grafik untuk menghindari tumpang tindih
    plt.legend(title='Features', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Mengatur layout agar elemen-elemen tidak saling tumpang tindih
    plt.tight_layout()

    # Menampilkan barplot
    plt.show()

# Contoh: Plot distribusi untuk cluster 0
plot_categorical_distribution_for_cluster('Features', cluster_number=0)
```

```python
custom_colors = [
    "#FF5733",  # Merah Cerah
    "#33FF57",  # Hijau Cerah
    "#3357FF",  # Biru Cerah
    "#FFFF33",  # Kuning Cerah
    "#FF33A1",  # Magenta Cerah
    "#33FFF7",  # Cyan Cerah
    "#FF8C00",  # Oranye Cerah
    "#8A33FF",  # Ungu Cerah
    "#FF33F6",  # Pink Cerah
    "#C4E17F",  # Hijau Pastel
    "#FFC300",  # Kuning Emas
    "#DAF7A6",  # Hijau Muda
    "#FFC0CB",  # Pink Pastel
    "#900C3F",  # Merah Tua
    "#581845",  # Merah Anggur
    "#DFFF00",  # Kuning Neon
]
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data yang sudah diberikan
df = data_model[["CAM", "CB", "CDM", "CF", "CM", "GK", "LB", "LM", "LW", "LWB", "RB", "RM", "RW", "RWB", "ST", "Clusters"]]

# Merapihkan data untuk barplot dengan menggunakan melt
df_melt = df.melt(id_vars='Clusters', var_name='Fitur', value_name='Nilai')

# Filter hanya untuk nilai 1
df_melt_1 = df_melt[df_melt['Nilai'] == 1]

# Menentukan warna untuk latar belakang berdasarkan cluster
background_colors = {
    cluster: color for cluster, color in zip(df['Clusters'].unique(), custom_colors)
}

# Membuat barplot
plt.figure(figsize=(12, 8))

# Menghitung posisi bar berdasarkan cluster
ax = sns.countplot(x='Clusters', hue='Fitur', data=df_melt_1, palette=custom_colors)

# Mengganti latar belakang untuk setiap cluster
# Mendapatkan posisi dan lebar setiap bar
bar_width = 0.8  # Lebar bar default
for i, cluster in enumerate(df['Clusters'].unique()):
    # Menggambar persegi panjang di belakang setiap cluster
    ax.add_patch(plt.Rectangle((i - bar_width/2, 0), bar_width, df_melt_1[df_melt_1['Clusters'] == cluster]['Fitur'].count(), color=background_colors[cluster], alpha=0.3))

# Menambahkan judul dan label
plt.title('Barplot Distribusi Nilai 1 Berdasarkan Cluster dengan Variabel Unik')
plt.xlabel('Cluster')
plt.ylabel('Jumlah')

# Menampilkan plot
plt.show()
```
